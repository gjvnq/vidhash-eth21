```python
from skimage.io import imread
from skimage.color import rgb2gray
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
```

```python
# implement 2D DCT
def dct2(img):
    from scipy.fftpack import dct
    return dct(dct(img.T, norm='ortho').T, norm='ortho')

# implement 2D IDCT
def idct2(img):
    from scipy.fftpack import idct
    return idct(idct(img.T, norm='ortho').T, norm='ortho')   
```

```python
img_original = rgb2gray(imread('../test-media/cat.png')) 
img_freq = dct2(img_original)
img_remade = idct2(img_freq)
np.allclose(img_original, img_remade)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.set_title('Original')
ax1.imshow(img_original, cmap='Greys')
ax2.set_title('Frequency Space')
ax2.imshow(img_freq)
ax3.set_title('Reconstructed')
ax3.imshow(img_remade)
```
